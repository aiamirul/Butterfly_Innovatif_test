# TASK 2
- FrameNo: frame position in the video
- (xmin,ymin): top-left pixel coordinate of the bounding box in the frame
- (xmax,ymax): bottom-right pixel coordinate of the bounding box in the frame
- Use “final.csv” to display the bounding box around all objects in the given video (Task2.avi).


```python
datafile = "final.csv"
dfPEDBOXES = pd.read_csv(datafile)
dfPEDBOXES.head()
dfPEDBOXES['label'].count()

#Clean Data to Integer
dfPEDBOXES['xmin'] = dfPEDBOXES.xmin.astype(int)
dfPEDBOXES['ymin'] = dfPEDBOXES.ymin.astype(int)
dfPEDBOXES['xmax'] = dfPEDBOXES.xmax.astype(int)
dfPEDBOXES['ymax'] = dfPEDBOXES.ymax.astype(int)
dfPEDBOXES['frameNo'] = dfPEDBOXES.frameNo.astype(int)
dfPEDBOXES
```

### Load Video and start drawing  save as output.avi 

```python
import numpy as np
import cv2

#LOAD VIDEO 
cap = cv2.VideoCapture('Task2.avi')

# Define the codec and create VideoWriter object
#fourcc = cv2.cv.CV_FOURCC(*'DIVX')
#out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.avi', fourcc, 10.0, (768,576))

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 768)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 576)
cap.set(cv2.CAP_PROP_FPS, 10)

#STREAM VIDEO BY FRAMES
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        #print(cap.get(cv2.CAP_PROP_POS_FRAMES))
        #FIND BOX FrameNo = VideoFrame
        boxes = dfPEDBOXES[dfPEDBOXES['frameNo'] == cap.get(cv2.CAP_PROP_POS_FRAMES)]
        
        #DRAW EACH BOX
        for row in boxes.iterrows():
            #Show some lines but not too much
            if(cap.get(cv2.CAP_PROP_POS_FRAMES) < 20):
                print("Drawing on Frame :",cap.get(cv2.CAP_PROP_POS_FRAMES),"Coordinates :",row[1].xmin,row[1].ymin,row[1].xmax,row[1].ymax)
            
            #image = cv2.rectangle(image, start_point, end_point, color, thickness)
            # represents the top left corner of rectangle
            start_point = (row[1].xmin, row[1].ymin)
            # represents the bottom right corner of rectangle
            end_point =(row[1].xmax, row[1].ymax)
            color = (0, 250, 0)
            frame = cv2.rectangle(frame, start_point, end_point, color, 5)
            
        # write the flipped frame
        out.write(frame)

        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Release everything if job is finished
cap.release()
out.release()
cv2.destroyAllWindows()
```

# TASK 3
- Using template matching, list all images in task3 folder if the image contains the given pattern (template.png).


```python
import cv2

def finditNowNoOverlap(file,templatefile):
    img_rgb = cv2.imread(file)
    template = cv2.imread(templatefile)
    h, w = template.shape[:2]

    res = cv2.matchTemplate(img_rgb, template, cv2.TM_CCOEFF_NORMED)

    threshold = 0.90
    loc = np.where( res >= threshold)

    matchfound = 0
    mask = np.zeros(img_rgb.shape[:2], np.uint8)
    for pt in zip(*loc[::-1]):
        if mask[pt[1] + int(round(h/2)), pt[0] + int(round(w/2))] != 255:
            mask[pt[1]:pt[1]+h, pt[0]:pt[0]+w] = 255
            matchfound += 1
            cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,255,0), 5)
            print("Found at Coordinate: ",res.shape)


    if(matchfound) :
        cv2.imwrite('found_map.jpg',img_rgb)
        print("Found {} matched in total!".format(matchfound))
        print(file)
    return matchfound
```

```python
import glob
import pandas as pd
Matched_list = []

images = 0;
templatefile = 'template.png'
for file in glob.glob("Task3/*"):
    print("Scanning: ",file)
    images = images + 1;
    found = finditNowNoOverlap(file,templatefile)
    if(found > 0):
        Matched_list.append(file)
```

```python
print("\n (Thresh > 90%) Matching Images :",len(Matched_list), " out of ", images ,' Images')
```
